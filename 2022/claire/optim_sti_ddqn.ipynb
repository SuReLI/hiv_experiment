# DQN

## DQN with target network

```python
from DDQN import DDQN
import torch

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
ddqn = DDQN(npatients=10,device=device)
r,s,a = ddqn.train()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#Plot loss over all batch
nlosses = len(ddqn.losses)
losses = [ddqn.losses[i].detach().numpy() for i in range(nlosses)]
plt.plot(losses)
```

```python
state_labels = ['T1','T1*','T2','T2*','V','E']
for i in range(len(state_labels)):
    plt.plot([t for t in range(200)],np.log10(s[-1][:,i]))
    plt.title(state_labels[i])
    plt.show()
```
